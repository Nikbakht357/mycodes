```python
df = pd.read_csv('D:/My Thesis/data in spss/imputed_Extra_tree .csv')
df = df.drop('ID' ,axis = 1)
```

```python
X=df.drop('Hospital_outcome',axis=1)
y=df['Hospital_outcome']
```

```python
categorical_features = ['Sex','Digestive_underlying', 'Respiratory_underlying','Infectious_underlying','Fatness_underlying', 'Urinary_underlying',
'Diabete_underlying', 'Highbloodpressure_underlying','Heart_underlying','Nervous_underlying', 'Blooddisease_underlying','Opiate', 'Education', 
'Urinary', 'Urinary_Incontinence', 'Previous_recurrent', 'Noconsious', 'GCS','Receivingblood', 'Conciousness', 'Patient_arrival', 'Time',
'Type_stroke', 'Treatment_method', 'Pneumonia', 'Convulsions','Individual', 'Motionless','Respiratory', 'Heart', 'Edema', 'Bloody',
'Vision_impairment', 'Sensory', 'Headache', 'Blurred_Vision','Reduce_strength_limb', 'Lackofbalance', 'Weakness', 'Delirium','Skin_problems',
'Aphasia_and_Dysarthria', 'Cranial_nerve_Disorder','Paralyzed', 'Digestive_problems']  # Replace with your categorical column names
numeric_features = ['Age','RR','HR','systolic','Lengthofhospitalization']
df[categorical_features] = df[categorical_features].astype('category')
```

```python
# Initial train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=101)
```

# OneHotEncoder and StandardScaler 

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
# Preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', StandardScaler(), numeric_features)
    ])
```

# IsolationForest

```python
from sklearn.ensemble import IsolationForest
# Outlier detection (on training data only)
outlier_detector = IsolationForest(contamination=0.05, random_state=101)
outliers = outlier_detector.fit_predict(preprocessor.fit_transform(X_train))
inlier_mask = outliers != -1
```

```python
# Apply outlier removal
X_train_clean, y_train_clean = X_train.iloc[inlier_mask], y_train.iloc[inlier_mask]
```

```python
# Preprocess cleaned training data
X_train_preprocessed = preprocessor.transform(X_train_clean)
# Evaluate on test set
X_test_preprocessed = preprocessor.transform(X_test)
```

```python
feature_names = preprocessor.get_feature_names_out()
```

# XGBoost Model

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
xgb_model = XGBClassifier(use_label_encoder=True, eval_metric='logloss',colsample_bytree=0.6,gamma=0.1,
learning_rate=0.1,max_depth=3, min_child_weight=5,n_estimators=100,subsample =1.0,random_state=101,verbose=1,n_jobs=-1)
xgb_model.fit(X_train_preprocessed, y_train_clean)
```

```python
y_pred = xgb_model.predict(X_test_preprocessed)
xgb_probs = xgb_model.predict_proba(X_test_preprocessed)[:, 1]  # Probabilities for the positive class (class 1)
```

```python
# Evaluate best model XGBoost
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.3f}")
```

# RandomForest Model

```python
###RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
# Initialize Random Forest
rf_model = RandomForestClassifier(bootstrap= False,max_depth= 30,max_features= 'sqrt',min_samples_split=2,min_samples_leaf=1,
                                  n_estimators= 150,oob_score =False,random_state=101,verbose=1,n_jobs=-1)
rf_model.fit(X_train_preprocessed, y_train_clean)
```

```python
# Evaluate best model XGBoost
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.3f}")
```

# DNN Model

```python
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import GridSearchCV
from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Reshape
def create_model(activation='relu'):
    model = Sequential([
        Reshape((X_train_preprocessed.shape[1],), input_shape=(X_train_preprocessed.shape[1],)),  # Explicit reshape
        Dense(64, activation=activation),
        Dense(32, activation=activation),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
dnn_model = KerasClassifier(
    model=create_model,
    activation='sigmoid',      # From best params
    batch_size=64,            # From best params  
    epochs=50, random_state=101,verbose=1              # From best params
)
dnn_model = dnn_model.fit(X_train_preprocessed, y_train_clean)
```

```python
y_pred = dnn_model.predict(X_test_preprocessed)
dnn_probs = dnn_model.predict_proba(X_test_preprocessed)[:, 1]  # Probabilities for the positive class (class 1)
```

```python
# Evaluate best model XGBoost
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.decomposition import PCA


svm_model = SVC(C=10,
    kernel='rbf',
    degree=1,
    gamma='auto',probability=True,verbose=1) 
svm_model.fit(X_train_preprocessed,y_train_clean)
```

```python
y_pred = svm_model.predict(X_test_preprocessed)
svm_probs = svm_model.predict_proba(X_test_preprocessed)[:, 1]  # Probabilities for the positive class (class 1)
```

```python
# Evaluate best model XGBoost
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.3f}")
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title('SVM Confusion Matrix')
plt.show()
```

```python
# 4. Decision Boundary Visualization (using PCA)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_preprocessed)

# Train a simplified SVM on PCA components
svm_pca = SVC(kernel='rbf', C=10, gamma='auto',probability=True)
svm_pca.fit(X_pca, y_train_clean)
```

```python
# Create mesh grid
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
# Plot decision boundary
DecisionBoundaryDisplay.from_estimator(
    svm_pca,
    X_pca,
    response_method="predict",
    alpha=0.3,
    cmap=plt.cm.RdYlBu,
    grid_resolution=100
)
# Scatter plot of actual data
plt.scatter(X_pca[y_train_clean==0, 0], X_pca[y_train_clean==0, 1], 
            c='blue', label='Negative Outcome', alpha=0.7, edgecolors='k')
plt.scatter(X_pca[y_train_clean==1, 0], X_pca[y_train_clean==1, 1], 
            c='red', label='Positive Outcome', alpha=0.7, edgecolors='k')

plt.title('SVM Decision Boundary (PCA Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# 5. Advanced: Probability Contours
Z = svm_pca.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.3, levels=20, cmap=plt.cm.RdYlBu)
plt.colorbar(label='Probability of Positive Outcome')
plt.scatter(X_pca[y_train_clean==0, 0], X_pca[y_train_clean==0, 1], 
            c='blue', label='Negative', alpha=0.5)
plt.scatter(X_pca[y_train_clean==1, 0], X_pca[y_train_clean==1, 1], 
            c='red', label='Positive', alpha=0.5)
plt.title('SVM Probability Contours')
plt.show()
```

# Initialize models

```python
# Get probabilities
models = {
    'XGBoost': xgb_model.predict_proba(X_test_preprocessed)[:, 1] ,
    'Random Forest': rf_model.predict_proba(X_test_preprocessed)[:, 1],
    'DNN': dnn_model.predict_proba(X_test_preprocessed)[:, 1],
    'SVM': svm_model.predict_proba(X_test_preprocessed)[:, 1]
}

plt.figure(figsize=(19,8))

for name, probs in models.items():
    # Compute ROC curve
    fpr, tpr, thresholds_roc = roc_curve(y_test, probs)
    g_means = np.sqrt(tpr * (1 - fpr))
    best_threshold_gmean = thresholds_roc[np.argmax(g_means)]
    
    # Compute Precision-Recall curve
    precisions, recalls, thresholds_pr = precision_recall_curve(y_test, probs)
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-8)
    best_threshold_f1 = thresholds_pr[np.argmax(f1_scores)]
    # Find the closest threshold to 0.5 in the ROC curve
    idx_roc = np.argmin(np.abs(thresholds_roc - 0.5))
    fpr_0_5, tpr_0_5 = fpr[idx_roc], tpr[idx_roc]

    # Find the closest threshold to 0.5 in the Precision-Recall curve
    idx_pr = np.argmin(np.abs(thresholds_pr - 0.5))
    recall_0_5, precision_0_5 = recalls[idx_pr], precisions[idx_pr]
    

    # Plot ROC Curve
    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f"{name}")
    plt.scatter(fpr[np.argmax(g_means)], tpr[np.argmax(g_means)], marker='*', color='black', s=65, label=f"{name} (Best G-Mean)")
    plt.scatter(fpr_0_5, tpr_0_5, marker='o', color='black', s=50, label=f"{name} (Threshold 0.5)")  # Circle marker at 0.5
    plt.plot([0, 1], [0, 1], linestyle='--', color='navy', label="No Skill")  # No-Skill baseline

# Plot Precision-Recall Curve
    plt.subplot(1, 2, 2)
    plt.plot(recalls, precisions, label=f"{name}")
    plt.scatter(recalls[np.argmax(f1_scores)], precisions[np.argmax(f1_scores)], marker='*', color='black', s=65, label=f"{name} (Best F1-Score)")
    plt.scatter(recall_0_5, precision_0_5, marker='o', color='black', s=50, label=f"{name} (Threshold 0.5)")  # Circle marker at 0.5
    no_skill = sum(y_test) / len(y_test)
    plt.plot([0, 1], [no_skill, no_skill], linestyle='--', color='navy', label="No Skill")  # No-Skill baseline

  # Formatting
    plt.subplot(1, 2, 1)
    plt.title("ROC Curve")
    plt.xlabel("1-specifity")
    plt.ylabel("Recall")
    plt.legend(fontsize=8, loc='lower right')
    plt.grid()

    plt.subplot(1,2,2)
    plt.title("Precision-Recall Curve")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend(fontsize= 8, loc='upper right')
    plt.grid()

plt.show()
```

```python
from imblearn.under_sampling import EditedNearestNeighbours
# Define resampling techniques
samplers = {
    "Original": (X, y),
    "RENN": RepeatedEditedNearestNeighbours(),
    "OSS": OneSidedSelection(),
    "SMOTE": SMOTE(),
    "SVM-SMOTE": SMOTE(sampling_strategy='minority', k_neighbors=5),
    "ENN-SMOTE": SMOTEENN()
}

# Apply resampling and store results
pca_results = {}
for name, sampler in samplers.items():
    if name == "Original":
        X_resampled, y_resampled = X, y
    else:
        X_resampled, y_resampled = sampler.fit_resample(X, y)
    
    # Apply PCA with 3 components
    pca = PCA(n_components=3)
    X_pca = pca.fit_transform(X_resampled)
    
    pca_results[name] = (X_pca, y_resampled)
```

```python
fig = plt.figure(figsize=(18, 12))

for i, (name, (X_pca, y_resampled)) in enumerate(pca_results.items(), 1):
    ax = fig.add_subplot(2, 3, i, projection='3d')
    
    scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y_resampled, cmap='coolwarm')
    
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.set_zlabel("PC3")
    ax.set_title(f" {name}")
    
    plt.colorbar(scatter, ax=ax)

plt.tight_layout()
plt.show()
```
